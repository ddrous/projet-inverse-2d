## Chargement des donnees

```python
import numpy as np
import pandas as pd
import ast
import matplotlib.pyplot as plt

from ast import literal_eval as l_eval
```

```python
types = {'rho_expr':str, 'E_u_expr':str, 'E_l_expr':str, 'E_d_expr':str, 'E_r_expr':str}
converters={'rho':l_eval, 'E_u':l_eval, 'F_u':l_eval, 'T_u':l_eval, 'E_d':l_eval, 'F_d':l_eval, 'T_d':l_eval, 'E_l':l_eval, 'F_l':l_eval, 'T_l':l_eval, 'E_r':l_eval, 'F_r':l_eval, 'T_r':l_eval}

df_0 = pd.read_csv("../../data/data/df_rho.csv", thousands=',', dtype=types, converters=converters)

# display(df_0.columns)
df_0.head(2)
```
